```python
import pandas as pd
```

```python
df = pd.read_csv('input.txt',delimiter=" ",header=None)
df
```

```python
df['char'] = df[1].apply(lambda x: x[0])

df['min'] = df[0].apply(lambda x: x.split('-')[0]).astype('int')
df['max'] = df[0].apply(lambda x: x.split('-')[1]).astype('int')
df = df.rename(columns={2:'password'})
del df[0]
del df[1]
df
```

```python
df['char_count'] = df.apply(lambda x: x['password'].count(x['char']),axis=1)
df
```

```python
len(df.where((df['char_count']>=df['min']) & (df['char_count']<=df['max'])).dropna())
```

```python
df = df.rename(columns={'min':'first','max':'second'})
del df['char_count']
df
```

```python
df['char_first'] = df.apply(lambda x: x['password'][x['first']-1] ,axis=1)
df['char_second'] = df.apply(lambda x: x['password'][x['second']-1],axis=1)
df
```

```python
len(df.where((df['char'] == df['char_first']) ^ (df['char'] == df['char_second'])).dropna())
```
